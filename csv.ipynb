# 爬虫数据存储

爬下来的数据需要存储起来，供之后使用

数据存储的形式可以是 

    1. CSV （少量数据）
    2. 数据库（大量数据）
  


## CSV

读取 Douban 图书 TOP250，数据存在CSV 文件里

    * 书名
    * 豆瓣链接
    * 作者
    * 出版社
    * 出版日期
    * 价格
    * 评分
    * 评分人数
    * 评论
    

```python
import requests
from lxml import etree
import csv
import re

headers = {
    'user-agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36"
}
```

```python
file = open("/Users/apple/Downloads/Code/crawler/douban250.csv",'wt', newline='', encoding= 'utf8')
writer = csv.writer(file)
writer.writerow(("name","url","auther","publisher","date","price","rate","rates_num","comments"))
```

```python
for i in range(0,250,25):      
    url = "https://book.douban.com/top250?start={}".format(str(i))
    res = requests.get(url, headers=headers)
    if res.status_code == 200:
        selector = etree.HTML(res.text)
        loop_points = selector.xpath('//tr[@class="item"]')
        name = loop_points[0].xpath('//td[2]/div[1]/a/@title')
        link = loop_points[0].xpath('//td[2]/div[1]/a/@href')
        details = loop_points[0].xpath('//td[2]/p[1]/text()')
        auther = [detail.split("/")[0].strip() for detail in details]
        publisher = [detail.split("/")[-3].strip() for detail in details]
        date = [detail.split("/")[-2].strip() for detail in details]
        price = [detail.split("/")[-1].strip() for detail in details]
        rate =  [point.xpath('td[2]/div[2]/span[2]/text()')  for point in loop_points]
        rates_num = [int(re.findall("\d+",x)[0]) for x in loop_points[0].xpath("//td[2]/div[2]/span[3]/text()")]
        comments = loop_points[0].xpath("//td[2]/p[2]/span/text()")
        
        for name, link, auther, publisher,date,price,rate,rates_num,comments in zip(name, link, auther, publisher,date,price,rate,rates_num,comments):
            print(name, link, auther, publisher,date,price,rate,rates_num,comments)
            writer.writerow((name, link, auther, publisher,date,price,rate,rates_num,comments))

    
file.close()
```
