# xpath

## 基本

Xpath路径可以通过 Chrome 复制得到

通过/text()可以获取标签中的文字信息

````python
import requests
from lxml import etree
````

````python


headers = {
    'user-agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36"
}

url = "https://bbs.hupu.com/31087594.html"
res = requests.get(url, headers=headers)
selector = etree.HTML(res.text)
id = selector.xpath('//*[@id="tpc"]/div/div[2]/div[1]/div[1]/a/text()')[0]
# /text() 可以获取标签中的文字信息。
print(id)
````

当需要进行用户囚的批量爬取时，通过类似于 BeautifulSoup 中的 selector() 方法删除谓语部分是不可行的。

这时的思路为“先抓大后抓小，寻找循环点” 。 

打开 Chrome 浏览器进行“检查”，通过“三角形符号”折叠元素，找到每个段子完整的信息标签

如图所示，每一个div标签为虎扑步行街一个回复的信息。



1. 首先通过复制构造 div标签路径，此时的路径为:

    `//div[@class=”floor“]`

这样就定位到了每个段子信息，这就是“循环点” 。

2. 通过 Chrome 浏览器进行“检查”定位用户 ID，复制 Xpath 到记事本中: 

    `//*[@id="104280"]/div[2]/div[1]/div[1]/a`

因为第一部分为循环部分，将其删除得到:

    `div[2]/div[1]/div[1]/a`

这便是用户ID的信息。

````python
url_infos = selector.xpath('//div[@class="floor"]')
len(url_infos)
for url_info in url_infos:
    name = url_info.xpath('//div[2]/div[1]/div[1]/a/text()')
print(name)

````

## 相同的字符开头的多个标签

有时候会遇到相同的字符开头的多个标签:

```html
    <li class=”tag-1”〉需要的内容 1</li> 
    <li class="tag-2’。需要的内容 2</li> 
    <li class=”tag-3”〉需要的内容 3</li>
```

想同时爬取时，不需要构造多个 Xpath 路径，通过 starts-with() 便可以获取多个标签内容:

Return all **div** elements in the XML document that have an attribute class whose attribute value starts with "tag".

````python
htmll = '''
<li class="tag-1"> a </li> 
<li class="tag-2"> b </li> 
<li class="tag-3"> c </li>
'''
selector = etree.HTML(htmll)
````

````python
contents = selector.xpath("//li[starts-with(@class,'tag')]/text()")
for content in contents:
    print(content)
````

## 标签套标签

当遇到标签套标签情况时 :
```html
<div class="red"> 需要的内容 1 <hl>需要的内容 2</hl> </div>
```

想同时爬取文本内容，可以通过 string(.)完成:

> The string function converts the given argument to a string.

https://developer.mozilla.org/en-US/docs/Web/XPath/Functions/string

````python
html2 = """

<div class="red-1"> 需要的内容 1 
    <hl> 需要的内容2 </hl> </div>

<div class="red-2"> 需要的内容 3 
    <hl> 需要的内容4 </hl> </div>
"""
selector = etree.HTML(html2)
````

````python
content1 = selector.xpath("//div[starts-with(@class,'red')]")[1]
content1.xpath("string(.)")

# Indicates the context node itself. It can be abbreviated as a single period (.).
````

# 性能对比

通过代码对正则表达式、 BeautifulSoup、 Lxml 进行性能对比。

https://bbs.hupu.com/31150516.html

爬取 Hupu 帖子回复的ID、发表时间、点亮数、文字内容


````python
import re
from bs4 import BeautifulSoup 
from lxml import etree
import time
````

````python
def re_scraper(res):
     # 尝试获取某个网页
    info_list = []
    if res.status_code == 200:
        ids = re.findall(
            '<a class="u" target="_blank" href="https://my.hupu.com/[0-9]*">(.*?)</a>', res.text, re.S)
        times = re.findall(
            '&nbsp;<span class="stime">(.*?)</span>', res.text, re.S)
        lights = re.findall(
            '亮了</a>\n\(<span class="stime">(.*?)</span>', res.text, re.S)

        # content没跑通
        contents = re.findall(
            '<table class="case" border="0" cellspacing="0" cellpadding="0">\n<tbody>\n<tr>\n<td>\n(.*?)</td>', res.text, re.S)

        for ida, time, light, content in zip(ids, times, lights, contents):
            info = {"id": ida,
                    "time": time,
                    "light": light,
                    "content": content}
            info_list.append(info)
        return info_list
````

````python
def bs_scraper(res):
    info_list = []
    if res.status_code == 200:
        soup = BeautifulSoup(res.text, "lxml")
        ids = [x.get_text().strip() for x in soup.select(
            "div > div.floor_box > div.author > div.left > a")]
        times = [x.get_text().strip() for x in soup.select(
            "div.floor_box > div.author > div.left > span.stime")]
        lights = [int(x.get_text().strip()) for x in soup.select(
            "div.floor_box > div.author > div.left > span.f444 > span > span")]
        contents = [x.get_text().strip() for x in soup.select(
            "div.floor_box > table > tbody > tr > td")]
        for ida, time, light, content in zip(ids, times, lights, contents):
            info = {"id": ida,
                    "time": time,
                    "light": light,
                    "content": content}
            info_list.append(info)
    return info_list
````

````python
def lxml_scraper(res):
    info_list = []
    if res.status_code == 200:
        selector = etree.HTML(res.text)
        loop_points = selector.xpath('//div[@class="floor"]')

        for loop_point in loop_points:
            ids = loop_point.xpath('//div[2]/div[1]/div[1]/a/text()')
            times = loop_point.xpath('//div[2]/div[1]/div[1]/span[2]/text()')
            lights = loop_point.xpath(
                '//div[2]/div[1]/div[1]/span[4]/span/span/text()')
            contents = [x.xpath("string(.)")for x in loop_point.xpath(
                '//div[2]/table/tbody/tr/td')]

        for ida, time, light, content in zip(ids, times, lights, contents):
            info = {"id": ida,
                    "time": time,
                    "light": light,
                    "content": content}
            info_list.append(info)

    return info_list
````

````python
if __name__ == '__main__':
    
    for name, scraper in [("regular expression", re_scraper), ("BeautifulSoup", bs_scraper), ("Lxml", lxml_scraper)]:
    
        urls = ["https://bbs.hupu.com/31087594-{}.html".format(str(i) for i in range(1,10))]
        start_time = time.time()
        for url in urls:
            res = requests.get(url, headers=headers)
            info_list = scraper(res)
        
        end_time = time.time()
        
        print(name, end_time-start_time)    
            
            
    
````
